```python
import matplotlib.pyplot as plt
data["category"].value_counts().plot.bar()
plt.show()
```

```python
data["duration"] = data["run_time"].astype(str).str[:3]
print(data["duration"].head(3))
```

```python
data["duration"] = data["duration"].astype(float)
data["duration"].head(2)
```

```python
d = data["duration"]
min = d.min()
max = d.max()
median = d.median()
mean = d.mean()
print(min, max, median, mean)

```
